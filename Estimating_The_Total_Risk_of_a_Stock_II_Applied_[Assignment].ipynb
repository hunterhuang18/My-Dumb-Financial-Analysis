```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = yf.download('AAPL', start='2000-01-01', end='2017-12-31')
df.head()
```

```python
def getStandardDev(df, annualization = False):
  '''
  type: df: Dataframe
        annualization : Boolean
  rtype: float

  parameters:
    df : Dataframe

    annunualization : If return the annual standard deviation
  '''
  if annualization == True:
    variance = np.var(df['Adj Close'], ddof=1)
    return np.sqrt(variance) * np.sqrt(252)
  else:
    variance = np.var(df['Adj Close'], ddof=1)
    return np.sqrt(variance)
```

```python
getStandardDev(df, annualization=True)[0]
```
