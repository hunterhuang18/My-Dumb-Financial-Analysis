```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
'''Function of expected returns'''

def getExpectedReturn(df, price, frequency = 'daily', rules = 'crude'):
  '''
  type: df: Dataframe
        price: str
        frequency: str
        rules: str
  rtype: float


  parameters:
    df : Dataframe

    price : Specific column of dataframe

    frequency :
       daily : Expected daily returns
       weekly : Expected weekly returns
       annual : Expected annual returns

    rules :
       crude : Arithmetic average method
       sophisticated : Geometric avergage method

  '''

  # Calculate the real daily return
  real_return = df[price].pct_change(1)

  # Calculate the expected daily retrun by mean method
  expected_daily_return = real_return.mean()

  # Calculate the expected weekly return
  if frequency == 'weekly':
    if rules == 'crude':
      expected_weekly_return = expected_daily_return * 5
      return expected_weekly_return
    elif rules == 'sophisticated':
      expected_weekly_return = (1 +expected_daily_return)**5-1
      return expected_weekly_return

    if frequency == 'annual':
      if rules == 'crude':
        expected_annual_return = expected_daily_return * 250
        return expected_annual_return
      elif rules == 'sophisticated':
        expected_annual_return = (1 +expected_daily_return)**250-1
        return expected_annual_return

  return expected_daily_return
```
